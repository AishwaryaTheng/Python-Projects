```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
car=pd.read_csv('cars.csv')
```

## Assignment1
**2. Generate a line plot graph for the column ‘model’ and ‘hp’:**
-->a. Map the ‘model’ column on the x-axis
-->b. Map the ‘hp’ column on the y-axis
-->c. Provide the x-axis label as Models of the cars
-->d. Provide the y-axis label as Horse-Power of Cars
-->e. Set the title as Model Names vs HorsePower

```python
## Generate a line plot graph for 'model' and 'hp'
plt.figure(figsize=(12, 6))
plt.plot(car['model'], car['hp'], marker='o', linestyle='-')
## Add labels and title
plt.xlabel('Models of the cars')
plt.xticks(rotation=90)
plt.ylabel('Horse-Power of Cars')
plt.title('Model Names vs HorsePower')
## Show the plot
plt.show()
```

## Assignment 2
**1. Generate a bar plot graph for the columns ‘carbs’ and ‘gear’:**
a. Map the ‘carbs’ onto the x-axis
b. Map the ‘gear’ onto the y-axis
c. Provide the x-axis label as Number of carburetors
d. Provide the y-axis label as Number of forward gears
e. Set the title as carbs vs gear


```python
plt.figure(figsize=(10, 6))
plt.bar(car['carb'], car['gear'], color='skyblue')

# Step 3: Add labels and title
plt.xlabel('Number of carburetors')
plt.ylabel('Number of forward gears')
plt.title('Carbs vs Gear')

# Step 4: Show the plot
plt.show()
```

## Assignment 3
***1. Plot a histogram for the column ‘wt’:
a. Map the ‘wt’ onto the x-axis
b. Provide the x-axis label as ‘weight of the cars’
c. Provide the y-axis label as ‘Count’

```python
plt.figure(figsize=(10, 6))
plt.hist(car['wt'], bins=30, color='skyblue', edgecolor='black')

# Step 3: Add labels and title
plt.xlabel('Weight of the cars')
plt.ylabel('Count')
plt.title('Histogram for the weight values')

# Step 4: Show the plot
plt.show()
```

## Assignment 4
**1. Plot a pie chart for columns: ‘cyl’ and ‘model’ form the mtcars.csv data
frame**

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)  # Create subplot for 'cyl'
car['cyl'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral', 'lightgreen'])
plt.title('Pie Chart for Cylinder Counts')

# Step 3: Plot a pie chart for 'model'
plt.subplot(1, 2, 2)  # Create subplot for 'model'
car['model'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightgreen', 'skyblue', 'orange'])
plt.title('Pie Chart for Model Counts')

# Step 4: Show the plot
plt.show()
```

```python
# Assuming df is your DataFrame
# If you don't have it yet, you should load or create it before running this code

# Step 1: Import necessary libraries
import matplotlib.pyplot as plt

# Step 2: Plot an area chart for 'am' and 'carb'
plt.figure(figsize=(10, 6))
plt.fill_between(car['am'], car['carb'], color='skyblue', alpha=0.4, label='Number of carburetors')

# Step 3: Add labels and title
plt.xlabel('Transmission')
plt.ylabel('Number of carburetors')
plt.title('Transmission vs Number of carburetors')
plt.legend()

# Step 4: Show the plot
plt.show()
```
